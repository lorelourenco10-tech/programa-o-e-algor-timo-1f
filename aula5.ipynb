```python
# Operadores aritiméticos

# Pedir ao usuário digitar 02 números

# Entrada

a = int (input("Digite um número: "))
b = int (input("Digite outro número: "))

# Processamento

# soma
# subtração
# multiplicação
# divisão
# divisão inteira
# resto
# potência

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b

# Saída

print("")
print("----------operadores aritiméticos----------")
print("-------------------soma--------------------")
print("-----------------subtração-----------------")
print("----------------multiplicação--------------")
print("------------------divisão------------------")
print("------------------div_int------------------")
print("---------------------resto-----------------")
print("------------------potência-----------------")
print("")
print(f"O resultado da soma de {a} mais {b} é igual a: {soma}")
print(f"O resultado da diferença de {a} menos {b} é igual a: {subtracao}")
print(f"O resultado da multiplicação de {a} vezes {b} é igual a: {multiplicacao}")
print(f"O resultado da divisão de {a} dividido por {b} é igual a: {divisao}")
print(f"O resultado da divisão inteira de {a} e {b} é igual a: {div_int}")
print(f"No resultado da divisão de {a} e {b} o resto é igual a: {resto}")
print(f"O resultado da potência de {a} e {b} é igual a: {potencia}")
```
