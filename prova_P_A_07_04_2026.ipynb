```python
# Programa que lê a idade do usuário

# Entrada

# > 17
# < 18
# < 60

idade = int (input("Digite a sua idade: "))

# Processamento e saída

if idade <= 17
  print (f"Você é menor de idade.")

elif idade > 18:
  print (f"Você é um adulto.")

else:
  print (f"Você é idoso(a).")
```

```python
# Crie um programa que leia o valor de uma compra e aplique o desconto conforme as regras:

# Entrada

# > R$100: sem desconto
# < R$100: 10% de desconto
# < R$500: 20% de desconto

valor = float (input("Digite o valor da sua compra: "))

# Processamento e saída

if valor <= 100:
  print (f"Você não terá nenhum desconto.")

elif valor < 100:
  print (f"Você terá 10% de desconto.")

else:
  print (f"Você terá 20% de desconto.")
```

```python
# Crie um programa que leia três valores e verifique se eles podem formar um triângulo.

# Entrada

lado1 = float (input("Digite o primeiro valor: "))
lado2 = float (input("Digite o segundo valor:  "))
lado3 = float (input("Digite o terceiro valor: "))

# Processamento e saída

valor1 = lado1 + lado2
valor2 = lado2 + lado3
valor3 = lado1 + lado3

if valor1 > lado3:
 print(f"Você tem um triângulo")

elif valor2 > lado1:
 print (f"Você tem um triângulo")

else:
  print (f"Você não tem um triângulo")

```

```python
# Crie um programa que leia a energia de um personagem (de 0 a 100) e classifique:

# Entrada

nome = str (input("Digite o nome do seu personagem: "))
energia = int (input("Digite a energia do seu personagem: "))

# = 0
# > 30
# < 30
# < 70

# Processamento e saída

if energia <= 0:
 print (f"Seu personagem {nome}, está morto")

elif energia <= 30:
  print (f"Seu personagem {nome}, está em estado crítico")

elif energia > 30:
  print (f"Seu personagem {nome}, está normal")

else:
 print (f"Seu personagem {nome}; está normal")
```

```python
# Fazer um programa em que calcula a média de notas de um estudante

# Entrada
nota1 = float (input("Digite a 1º nota :"))
nota2 = float (input("Digite a 2º nota :"))
nota3 = float (input("Digite a 3º nota :"))
nota4 = float (input("Digite a 4º nota :"))

# Processamento e saída

# Média maior ou igual a 7: Aprovado
# Média entre 5 e 6.9: Recuperação

media = nota1 + nota2 + nota3 + nota4
media2 = media / 4

if media2 >= 7:
 print (f" Você foi aprovado")

elif media2 <= 6.9:
 print (f" Você está de recuperação")

else:
  print (f"Você foi reprovado")
```

```python
# Crie um programa que leia o salário de um funcionário e sua pontuação de desempenho.

# Entrada

# Pontuação menor que 5: sem bônus
# Pontuação de 5 à 7: bônus de 5% do salário
# Pontuação de 8 à 9: bônus de 10% do salário
# Pontuação igual à 10: bônus de 15% do salário

nome = str (input("Digite seu nome: "))
salario = int (input("Digite seu salário: "))
pontuacao = int (input("Digite sua pontuação deste mês: "))

porcentagem1 = (5 / 100) * salario
porcentagem2 = (10 / 100) * salario
porcentagem3 = (15 / 100) * salario

# Processamento e saída

if pontuacao < 5:
  print (f"Que pena! Esse mês você não terá nenhum bônus")

elif pontuacao <= 7:
  print (f"Esse mês você terá 5% de bônus, ou seja ganhará {porcentagem1} a mais no seu salário deste mês")

elif pontuacao <= 9:
  print (f"Esse mês você terá 10% de bônus, ou seja ganhará {porcentagem2} a mais no seu salário deste mês")

else:
  print (f"Parabéns! Você ganhou uma porcentagem de 15% a mais no seu salário, o que significa que ganhará {porcentagem3} a mais no salário deste mês")
```
