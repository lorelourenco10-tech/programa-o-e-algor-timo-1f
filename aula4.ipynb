```python
# Entrada
lado1 = float (input("Digite o comprimento:"))
lado2 = float (input("Digite a altura:"))

# Processamento

area = lado1 * lado2

# Saída

print (f"O total da área é :{area} metros quadrados")
```

```python
# Entrada
numero1 = float (input("Digite o número :"))

# Processamento

dobro = numero1 * 2

# Saída

print (f"O dobro do número é :{dobro}")
```

```python
# Entrada
nota1 = float (input("Digite a 1º nota :"))
nota2 = float (input("Digite a 2º nota :"))
nota3 = float (input("Digite a 3º nota :"))
nota4 = float (input("Digite a 4º nota :"))

# Processamento

media = nota1 + nota2 + nota3 + nota4
media2 = media / 4

# Saída

print (f"O resultado da média das notas é :{media2}")
```

```python
# Entrada
nome = str (input("Digite seu nome completo : "))
idade = int (input("Digite sua idade : "))

# Saída

print (f"Olá {nome}, você tem {idade} anos")
```
