```python
# Escrever um programa em que o usuário digita um número e o programa imprime até o número digitado

numero = int (input("Digite um número inteiro: "))

for i in range(1, numero + 1):
  print (i)
```

```python
# Escrever um programa em que o usuário digita um número e o programa faz a contagem regressiva

numero = int(input("Digite um número para iniciar a contagem regressiva: "))

# Começa em um número, vai até -1 (para incluir o 0) e subtrai 1 a cada volta
for i in range (numero, -1, -1):
  print (i)
```
