```python
# Escrever um programa em que o usuario digita um número de 1 à 10 e o programa imprime a sua tabuada.

numero = int (input("Digite um número de 1 à 10. "))

print (f"Tabuada do {numero}: ")

for i in range (1, 11):
  resultado = numero * i
  print (f" {numero} x {i} = {resultado}")
```
