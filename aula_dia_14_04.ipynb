```python
# Peça um número ao usuário e informe se ele é par ou ímpar.

# Entrada

numero = int (input("Digite um número: "))

# Processamento e saída

if numero % 2 == 0:
  print (f"O número {numero} é um número par.")

else:
  print (f"O número {numero} é um número impar.")

```

```python
# Solicite dois números e mostre qual é o maior (ou se são iguais).

# Entrada
numero1 = int (input("Digite um número: "))
numero2 = int (input("Digite outro número: "))

# Processamento e saída

if numero1 > numero2:
  print (f"O {numero1} é maior do que o {numero2}.")

elif numero2 > numero1:
  print (f"O {numero2} é maior que o {numero2}.")

else:
  print (f"Os números {numero1} e {numero2} são iguais.")
```

```python
# Peça a nota de um aluno e informe:

# Entrada

# Aprovado (nota ≥ 7)
# Recuperação (5 ≤ nota < 7)
# Reprovado (nota < 5)

nota = float (input("Digite sua nota: "))

# Processamento e saída

if nota >= 7:
  print (f"Você foi aprovado!")

elif nota >= 5 and nota < 7:
  print (f"Você está de recuperação.")

else:
  print (f"Você foi reprovado.")
```

```python
# Receba a idade e classifique:

# Entrada

# Criança (0–12)
# Adolescente (13–17)
# Adulto (18–59)
# Idoso (60+)

idade = int (input("Digite sua idade: "))

# Processamento e saída

if idade <= 12:
  print (f"Você é uma criança.")

elif idade <= 17:
  print (f"Você é um adolescente.")

elif idade <= 59:
  print (f"Você é um adulto.")

else:
  print (f"Você é um idoso.")
```

```python
# Peça dois números e uma operação (+, -, *, /).
# Entrada

numero1 = int (input("Digite um valor: "))
numero2 = int (input("Digite um segundo valor:"))
operacao = str (input("Escreva uma operação que gostaria realizar com esses dois valores (sendo + soma, - subtração, * multiplicação, / divisão): "))
soma = numero1 + numero2
subtracao = numero1 - numero2
multiplicacao = numero1 * numero2
divisao = numero1 / numero2

# Processamento e saída

if operacao == "+":
  print (f"Sua soma dá o valor de {soma}.")

elif operacao == "-":
  print (f"Sua subtração dá o valor de {subtracao}")

elif operacao == "*":
  print (f"Sua multiplicação dá o valor de {multiplicacao}")

else:
  print (f"Sua divisão dá o valor de {divisao}")
```

```python
# Peça um número de 1 a 7 e mostre o dia correspondente (ex: 1 = domingo).

# Entrada
numero = int (input("Digite um número de 1 à 7: "))

# Processamento e saída

if numero == 1:
  print (f"O dia da semana que corresponde ao número digitado é domingo. ")

elif numero == 2:
  print (f"O dia da semana que corresponde ao número digitado é segunda- feira.")

elif numero == 3:
  print (f"O dia da semana que corresponde ao número digitado é terça- feira. ")

elif numero == 4:
  print (f"O dia da semana que corresponde ao número digitado é quarta- feira. ")

elif numero == 5:
  print (f"O dia da semana que corresponde ao número digitado é quinta- feira. ")

elif numero == 6:
  print (f"O dia da semana que corresponde ao número digitado é sexta- feira. ")

else:
  print (f"O dia da semana que corresponde ao número digitado é sábado. ")
```

```python
# Ano bissexto
# Solicite um ano e verifique se é bissexto ou não

# Entrada

ano = int (input("Digite um ano: "))

# Definição da função para verificar ano bissexto
def eh_bissexto(ano):
    return (ano % 400 == 0) or (ano % 4 == 0 and ano % 100 != 0)

# Processamento e saída

if eh_bissexto(ano):
    print(f"{ano} é bissexto.")

else:
  print (f"Esse ano não é nem será bissexto.")
```

```python
# Desconto em compras

# Entrada

# >= 100: 10% de desconto
# >= 50 e <100: 5% de desconto
# < 50: sem desconto

valor = float (input("Digite o valor da sua compra: "))

porcentagem1 = (10 / 100) * valor
porcentagem2 = (5 / 100) * valor

# Processamento e saída

if valor >= 100:
  print (f"Você terá 10% de desconto, {porcentagem1} de desconto na sua compra.")

elif valor >= 50:
  print (f"Você terá 5% de desconto, {porcentagem2} de desconto na sua compra.")

else:
  print (f"Você não terá nenhum desconto nessa compra.")
```
